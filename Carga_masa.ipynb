```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def funcion_ajuste(x, a, b):
    return a * x + b
Corriente = np.array([0.35,0.5,0.7,1,1.3,1.6,2,2.11,2.4,2.7,3,3.3,3.6,3.9,4.14])
Campo_magnetico = np.array([0.2,0.4,0.5,0.8,1,1.3,1.6,1.7,1.9,2.2,2.4,2.6,2.9,3.1,3.3])

popt, pcov = curve_fit(funcion_ajuste, Corriente, Campo_magnetico)
pendiente, intercepto = popt
perr = np.sqrt(np.diag(pcov))

incertidumbre_pendiente = perr[0]
incertidumbre_intercepto = perr[1]

print(f"Pendiente: {pendiente:.4f} ± {incertidumbre_pendiente:.4f}")
print(f"Intercepto: {intercepto:.4f} ± {incertidumbre_intercepto:.4f}")

x_ajustado = np.linspace(min(Corriente), max(Corriente), 100)
y_ajustado = funcion_ajuste(x_ajustado, pendiente, intercepto)

yerr = np.sqrt(pcov[0,0]*Corriente**2 + pcov[1,1] + 2*pcov[0,1]*Corriente)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 10), sharex=True)

ax1.errorbar(Corriente, Campo_magnetico, yerr=yerr, fmt='o', color='hotpink')
ax1.plot(x_ajustado, y_ajustado, color='plum', linestyle="--")
ax1.set_ylabel("Campo inducido B")
ax1.set_title("Campo magnético en función de la corriente")
ax1.grid(False)

residuals = Campo_magnetico - funcion_ajuste(Corriente, pendiente, intercepto)
ax2.plot(Corriente, residuals, marker="*", linestyle=" ", color="palevioletred")
ax2.axhline(y=0, color="gray", linestyle="--")
ax2.set_xlabel("Corriente A")
ax2.set_ylabel("Residuales")
ax2.grid(False)

plt.tight_layout()
plt.show()
```
