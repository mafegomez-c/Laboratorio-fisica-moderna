```python
import numpy as np
import matplotlib.pyplot as plt


dt= 0.01
innte= 10000

#aqui se normaliza la constante de la fuerza para simplificar calculos

k_delta_q_sobre_m= 1

#aqui condiciones iniciales que seran consideradas, considerar que s epueden conseguir parametros b arbitrarios

x0= -100

parametrosb_y0= [-3,-1,0,1,2,5]

vx0= 10.0
vy0= 0.0

#función para calcular el ángulo de dispersión

def calcular_theta(vx, vy):
    return np.degrees(np.arctan2(vy, vx))

# se crea función de ruutherford

def rutherford(x0,vx0,vy0, dt):

  plt.figure(figsize=(8, 6))


  #se toman cada uno de los parametros b escogidos como y0

  for b in parametrosb_y0:

    x= x0
    y= b
    vx=vx0
    vy=vy0
    rmin= np.inf

    # aqui se guardara la historia de trayectoria cuando se actualizen posiciones en x, y
    trayectoria_x= []
    trayectoria_y= []

    for i in range(innte):
      r= np.sqrt(x**2 + y**2)
      if r < 0.1:
        break
      if r < rmin:
        rmin= r

      # componentes de la aceleración tomando cte de fuerza normalizada

      ax= k_delta_q_sobre_m * x / r**3
      ay= k_delta_q_sobre_m * y / r**3

      # se actualiza velocidad y posición para considerar la trayectoria

      vx += ax * dt
      vy += ay * dt

      x += vx * dt
      y += vy * dt

      trayectoria_x.append(x)
      trayectoria_y.append(y)

    theta= calcular_theta(vx, vy)
    plt.plot(trayectoria_x, trayectoria_y, label=f"b = {b}")
    print(f"Para b = {b}, r_min = {rmin:.2f}, θ = {theta:.2f}°")

  plt.axhline(0, color='k', linestyle='--', linewidth=0.5)
  plt.axvline(0, color='k', linestyle='--', linewidth=0.5)
  plt.scatter([0], [0], color='red', s=100, label='Núcleo')
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.title("Simulación de Dispersión de Rutherford")
  plt.legend()
  plt.grid(True)
  plt.show()

rutherford(x0, vx0, vy0, dt)
```

```python
import numpy as np
import matplotlib.pyplot as plt

dt = 0.01
innte = 10000
k_delta_q_sobre_m = 1

x0 = -100
parametrosb_y0 = np.array([-3, -1, 1, 2, 3,4,5,6,7])
vx0 = 10.0
vy0 = 0.0

def calcular_theta(vx, vy):
    return np.degrees(np.arctan2(vy, vx))

lista_theta = []
lista_rmin = []

def rutherford(x0, vx0, vy0, dt):
    plt.figure(figsize=(6, 10))

    for b in parametrosb_y0:
        x = x0
        y = b
        vx = vx0
        vy = vy0
        rmin = np.inf
        trayectoria_x = []
        trayectoria_y = []

        for i in range(innte):
            r = np.sqrt(x**2 + y**2)
            if r < 0.1:
                break
            if r < rmin:
                rmin = r

            ax = k_delta_q_sobre_m * x / r**3
            ay = k_delta_q_sobre_m * y / r**3

            vx += ax * dt
            vy += ay * dt
            x += vx * dt
            y += vy * dt

            trayectoria_x.append(x)
            trayectoria_y.append(y)

        theta = calcular_theta(vx, vy)
        lista_theta.append(theta)
        lista_rmin.append(rmin)

        plt.plot(trayectoria_x, trayectoria_y, label=f"b = {b}")

    plt.axhline(0, color='k', linestyle='--', linewidth=0.5)
    plt.axvline(0, color='k', linestyle='--', linewidth=0.5)
    plt.scatter([0], [0], color='red', s=100, label='Núcleo')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Simulación de Dispersión de Rutherford")
    plt.legend()
    plt.grid(True)
    plt.show()

rutherford(x0, vx0, vy0, dt)

plt.figure()
plt.plot(parametrosb_y0, lista_theta, color='palegreen', label="Ángulo de dispersión",marker='o', markerfacecolor="salmon", markeredgecolor="salmon")
plt.ylabel("Ángulo de dispersión θ (grados)")
plt.title("Relación entre b y θ en la dispersión de Rutherford")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
plt.plot(parametrosb_y0, lista_rmin, color='skyblue',marker='o', markerfacecolor="hotpink", markeredgecolor="hotpink", label="Distancia mínima al núcleo")
plt.xlabel("Parámetro de impacto b")
plt.ylabel("Distancia mínima r_min")
plt.title("Relación entre b y r_min en la dispersión de Rutherford")
plt.legend()
plt.grid(True)
plt.show()
#esta es la prueba con valores de b incluyendo negativos
```

```python
deft = 0.01
innte = 10000
k_delta_q_sobre_m = 1
x0 = -100
parametrosb_y0 =  [1, 2, 3, 4, 5]
vx0 = 10.0
vy0 = 0.0

def calcular_theta(vx, vy):
    return np.degrees(np.arctan2(vy, vx))

lista_b = []
lista_theta = []
lista_rmin = []

def rutherford(x0, vx0, vy0, dt):
    plt.figure(figsize=(6, 10))

    for b in parametrosb_y0:
        x, y = x0, b
        vx, vy = vx0, vy0
        rmin = np.inf
        historiar= []
        ind= []
        contador = 0
        trayectoria_x, trayectoria_y = [], []

        for _ in range(innte):
            r = np.sqrt(x**2 + y**2)
            historiar.append(r)
            contador +=1
            ind.append(contador)
            if r < 0.1:
                break
            if r < rmin:
                rmin = r
            ax = k_delta_q_sobre_m * x / r**3
            ay = k_delta_q_sobre_m * y / r**3


            vx += ax * dt
            vy += ay * dt
            x += vx * dt
            y += vy * dt

            trayectoria_x.append(x)
            trayectoria_y.append(y)

        theta = calcular_theta(vx, vy)
        plt.plot(trayectoria_x, trayectoria_y, label=f"b = {b}")
        print(f"Para b = {b}, r_min = {rmin:.2f}, θ = {theta:.2f}°")

        lista_b.append(b)
        lista_theta.append(theta)
        lista_rmin.append(rmin)
        plt.plot(ind,historiar, label=f"distancia para r={b}", color="darkcyan")
        plt.xlabel("Indice del punto")
        plt.ylabel("Distancia R ")
        plt.show()
    plt.axhline(0, color='k', linestyle='--', linewidth=0.5)
    plt.axvline(0, color='k', linestyle='--', linewidth=0.5)
    plt.scatter([0], [0], color='red', s=100, label='Núcleo')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Simulación de Dispersión de Rutherford")
    plt.legend()
    plt.grid(True)
    plt.show()

rutherford(x0, vx0, vy0, deft)

plt.figure(figsize=(7, 5))
plt.plot(lista_b, lista_theta, 'plum', label='Simulación', marker='o', markerfacecolor="lightseagreen", markeredgecolor="lightseagreen")
plt.xlabel("Parámetro de impacto b")
plt.ylabel("Ángulo de dispersión θ (grados)")
plt.title("Ángulo de dispersión θ vs b")
plt.grid(True)
plt.legend()
plt.show()



#graficamos individualmente para observar el comportamiento de r_min respecto a b
```
