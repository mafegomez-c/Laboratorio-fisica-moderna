```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def funcion_ajuste(x, a, b):
    return a * x + b

distancia = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70])
num_minimos = np.array([1, 10, 20, 35, 50, 65, 80, 90, 106, 124, 141, 158, 173, 188, 205])

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 8), sharex=True,
                               gridspec_kw={'height_ratios': [3, 1]})

ax1.plot(distancia, num_minimos, marker="o", label="Distancia", color="hotpink")

ax1.set_xlabel("Distancia (μm)")
ax1.set_ylabel("Número de Mínimos")
ax1.set_title("Distancia vs Número de Mínimos")
ax1.legend()
ax1.grid(True)

x_ajustado = np.linspace(min(distancia), max(distancia), 100)
popt, pcov = curve_fit(funcion_ajuste, distancia, num_minimos)
a, b = popt
y_ajustado = funcion_ajuste(x_ajustado, a, b)
ax1.plot(x_ajustado, y_ajustado, linestyle="--", label="Ajuste lineal")

equation = f"y = {a:.2f}x + {b:.2f}"
ax1.text(0.05, 0.95, equation, transform=ax1.transAxes, fontsize=12,
         verticalalignment='top')

residuals = num_minimos - funcion_ajuste(distancia, a, b)
ax2.plot(distancia, residuals, marker='*', linestyle=' ', label='Residuales', color='darkcyan')
ax2.set_xlabel("Distancia (μm)")
ax2.set_ylabel("Residuales")
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def funcion_ajuste(x, a, b):
    return a * x + b

distancia = np.array([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70])
num_minimos = np.array([1, 10, 20, 35, 50, 65, 80, 90, 106, 124, 141, 158, 173, 188, 205])

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 8), sharex=True,
                               gridspec_kw={'height_ratios': [3, 1]})

ax1.plot(distancia, num_minimos, marker="o", label="Distancia", color="hotpink")

ax1.set_xlabel("Distancia (μm)")
ax1.set_ylabel("Número de Mínimos")
ax1.set_title("Distancia vs Número de Mínimos")
ax1.legend()
ax1.grid(True)

x_ajustado = np.linspace(min(distancia), max(distancia), 100)
popt, pcov = curve_fit(funcion_ajuste, distancia, num_minimos)
a, b = popt
y_ajustado = funcion_ajuste(x_ajustado, a, b)
ax1.plot(x_ajustado, y_ajustado, linestyle="--", label="Ajuste lineal")

equation = f"y = {a:.2f}x + {b:.2f}"
ax1.text(0.05, 0.95, equation, transform=ax1.transAxes, fontsize=12,
         verticalalignment='top')

residuals = num_minimos - funcion_ajuste(distancia, a, b)
ax2.plot(distancia, residuals, marker='*', linestyle=' ', label='Residuales', color='darkcyan')
ax2.set_xlabel("Distancia (μm)")
ax2.set_ylabel("Residuales")
ax2.grid(True)

plt.tight_layout()

uncertainty_a = np.sqrt(pcov[0, 0])
uncertainty_b = np.sqrt(pcov[1, 1])

print(f"Incertidumbre de la pendiente (a): {uncertainty_a:.4f}")
print(f"Incertidumbre del punto de corte (b): {uncertainty_b:.4f}")

plt.show()
```
